```python
# ライブラリのインポート
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
# データの準備
totsuka_kodate = pd.read_csv("totsuka_kodate.csv")
train_x = pd.get_dummies( totsuka_kodate[['築年数','最寄駅距離（分）','延床面積（㎡）','面積（㎡）','前面道路幅員（ｍ）','地区名', '前面道路種類', '土地の形状']])
train_y = totsuka_kodate['取引価格']
train_x, val_x, train_y, val_y = train_test_split(train_x, train_y, train_size=0.7, test_size=0.3, random_state=0)
# モデルの定義
model = RandomForestRegressor(n_estimators=500, max_depth=9, max_features=25, random_state=0)
# 学習
model.fit(train_x,train_y)
# 評価
print(f'訓練データ精度：{model.score(train_x, train_y)}')
print(f'検証データ精度：{model.score(val_x, val_y)}')
```
