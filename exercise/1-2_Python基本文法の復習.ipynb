## 変数と型
Pythonでは、**変数**を使用する前に**型**の記述をする必要がありません。  
以下のように、変数に値を代入することができます。  

```python
a = 123
# google colaboratory(jupyter notebook)では、最後に実行した内容が画面に表示できる場合には、自動で画面表示される。
a
```

Pythonでは型の宣言は必要ありませんが、格納されたものにより**変数の型が自動的に判別**されます。
Pythonの主な型を以下に示します。

また、**#はコメントを表し**、同じ行で#以降に書かれた内容はコードとして認識されません。

```python
a = 123             # 整数型（int）
b = 123.456         # 浮動小数点型（float）
c = "Hello World!"  # 文字列型（str）、文字列の記述には""と''どちらも使える。
d = True            # 論理型（bool）
e = [1, 2, 3]       # リスト型（list）
```

## 演算子
Pythonの**演算子**を紹介します。

**print関数**を使うと文字列や変数の内容を**画面に表示**することができます。

```python
a = 3
b = 4

# 足し算
c = a + b
print(c)

# 引き算
d = a -b
print(d)

# 掛け算
e = a * b
print(e)

# 割り算
f = a / b
print(f)
```

## リスト
**リスト**は、複数の値をまとめて扱う場合に使用します。  
リストは**全体を`[]`で囲み、各要素は`,`で区切ります**。  
Pythonのリストはどのような型の値でも格納することができ、リストの中にリストを格納することもできます。  
リストの各要素へのアクセスは**インデックス**を使い、要素の追加や入れ替えなどが可能です。

```python
# リストの作成
mylist = [1, 2, 3, 4, 5]

# 3番目の要素を取得する
# (インデックスは0から始まるため、0,1,2と指定する)
a = mylist[2]
print(a)

# 3番目の要素の値を変更する
mylist[2] = 7
print(mylist) # リスト自体を指定して、内容を表示することもできる
```

また、**「スライス」**を使用すると範囲指定をして要素を取得することができます。  
**mylist[start:end]**とすることでstartからendの手前までの要素を指定することができます。

```python
# 以下の指定でインデックスが1と2の要素を取り出すことができます
mylist[1:3]
```

リストは2次元以上の値を扱うこともできます。

```python
# 2次元リストの作成
mylist2 = [[1,2,3],
           [4,5,6]]
# 2次元以上のリストの要素を指定するときは、それぞれの次元のインデックスを指定します
print(mylist2[0][1])
print(mylist2[1][2])
```

## 辞書
**辞書**は、**キー**と**値**の組合せでデータを格納します。　  
次の例は、**文字列をキーとして辞書を作成**し、値の取得や入れ替え、要素の追加を行なっています。

```python
# 辞書の作成
mydictionary = {"Apple":"りんご", "Peach":"もも"}
# キーに"Apple"を指定して値を取得
print(mydictionary["Apple"])

# 値を変更する
mydictionary["Apple"] = "100円"
print(mydictionary["Apple"])

# 要素を追加する
mydictionary["Pear"] = "梨"
mydictionary
```

## if文  
**分岐**には**if文**を用います。  
ifの条件が満たされていなければ、**elif**の条件が上から順番に判定されます。  
これらの条件がすべて満たされていなければ、**else**内の処理が実行されます。  
pythonではif文の後ろの **:** と**インデント**を用いてif文を記載します。

```python
temperature = 36.5

if temperature < 37:
  print("仕事に行く")
elif temperature < 37.5:
  print("仕事を休む")
else:
  print("病院へ行く")
```

## for文
指定した回数**ループ**するためには**for文**を用います。  
ループする**範囲を指定**するためには、リストやrangeを**in**とともに用います。  

```python
# リストをループで処理する場合
sum = 0

for number in [1,3,10]:
  print(number)
  sum += number # +=は左辺の数値と右辺の数値を足し算した結果を左辺に代入します

print("---")
print(sum)
```

```python
# rangeを使ってループする場合は以下のように指定する。[]部分は省略可能で、省略すると0から指定した終了番号まで繰り返される。
# range([開始番号,] 終了番号［, ステップ数])
for i in range(3):
   print(i)
```

## 関数
**関数**を用いることで、複数行の処理をまとめることができます。  
関数は**def**のあとに**関数名**を記述し、**()の中に引数を記述**します。  
**return**のあとの値が**戻り値**になります。

```python
# 関数を定義
def add(a, b):
  sum = a + b
  return sum

# 引数を渡して関数を実行
print(add(10, 20))

# 以下のような指定の仕方もできる
print(add(a=1, b=2))
```

**引数にはデフォルト値を設定できます**。  
デフォルト値を設定すると、関数を呼び出す際にその引数を**省略**できます。  
以下の例では、第2引数にデフォルト値が設定されています。  

```python
def add(a=5, b=10):
  sum = a + b
  return sum

# 第一引数を指定しないで実行
add(b=2)
```

## クラス
本コースで必要なプログラミングスキルとして目安にしている「Python1」コースで扱っていない内容ですが、知っていた方が理解がはかどる為、簡単にクラスについて説明します。<br><br>
Pythonでは、オブジェクト指向プログラミングが可能です。  
オブジェクト指向では、**クラス**と**インスタンス**という概念があります。  
**クラスはひな形**のようなもので、**インスタンスはそれを実体化したもの**です。  
クラスから複数のインスタンスを生成することができます。  

オブジェクト指向をよく知らないという場合は、複数の関連する変数や関数をまとめて管理するために使用するものと考えてください。
<br><br>

以下の例では、**Manというクラス**を定義し、クラスに**\_\_init\_\_メソッド**、**mynameメソッド**、**yournameメソッド**が実装されています。

<br>
※ 関数とメソッドはどちらも複数の処理をまとめて呼び出せるようにしたものです。一般にクラス内で定義した関数をメソッドと呼びますが、本コースでは違いを意識しなくても問題ありません。

```python
class Man:
  def __init__(self, name):
    self.name = name
  
  def myname(self):
    print("My name is " + self.name)

  def your_name(self, yourname="unknown"):
    print("Your name is " + yourname)
```

Pythonのメソッドは**第1引数にself（インスタンス）を受け取る**という特徴があります。  
インスタンスの持つ変数（インスタンス変数）にアクセスするにはself.変数名でアクセスします。    

　**\_\_init\_\_**は特殊なメソッドで、インスタンスの初期設定を行います。  
 上記のクラスでは、**self.name = name**で引数として受け取った値をインスタンス変数**self.name**に代入します。

上記のクラス**Man**から、以下のように**インスタンスを生成しメソッドを呼び出す**ことができます。  

```python
# name = "Tanaka Taro"でインスタンスを生成
taro = Man("Tanaka Taro")
# mynameメソッドを呼び出す
taro.myname()
```

```python
taro.your_name("Suzuki Jiro")
```

また、クラスには継承という機能があります。  
クラスを継承することで、既存のクラスを引き継いで新たなクラスを定義することができます。  
以下の例では、**Man**クラスを継承して**SuperMan**クラスを定義しています。

```python
class SuperMan(Man):
  def fly(self):
    print("I can fly")
```

**fly**メソッドを新たに追加しました。  
**SuperManクラス**から**インスタンスを生成**し、**メソッド**を呼び出してみましょう。  

継承元の**Manクラス**で定義されたメソッドは、これを継承した**SuperManクラス**のインスタンスからも、同じように呼び出すことができます。  
クラスの継承を利用すれば、**複数のクラスの共通部分を継承元のクラスにまとめる**ことができます。  

```python
saburo = SuperMan("Sato Saburo")
saburo.myname()
saburo.fly()
```
