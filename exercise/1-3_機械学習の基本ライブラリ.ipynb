## ①NumPy
- Pythonベースの科学計算の基本ライブラリ
  - **ベクトル**や**行列**などの**多次元配列の計算**
  - **NumPy配列（np.array）**というデータの扱い方が、機械学習のスタンダード
  - その他、**乱数生成**など機能多数


```python
# 行列の内積計算
import numpy as np

X = np.array([ [1, 2], [3, 4] ]) # numpy配列Xを作成
Y = np.array([ [5, 6], [7, 8] ]) # numpy配列Yを作成
Z = np.dot(X,Y)

print (Z)
```

## ③matplotlib
- データを**プロット**した画像を生成するPythonライブラリ
  - データや計算結果を**描画**して視覚的に状態を確認できる
  - **散布図**や**ヒストグラム**、**パイチャート**など多彩な描画が可能
  - グラフ描画ライブラリの**seaborn**は**matplotlib**をベースに作成されており、matplotlibと同様に機械学習で良く使われる


```python
# シグモイド関数の描画
import numpy as np
import matplotlib.pyplot as plt

# シグモイド関数のx,yの値をそれぞれリストで用意する
x = np.arange(-7, 7, 0.1)
y = 1 / (1 + np.exp(-x) )

plt.plot(x, y)        # x,yを曲線でプロット
plt.title("sigmoid")  # タイトルの設定
plt.grid()            # グリッド線の表示
plt.show()            # グラフの表示
```
